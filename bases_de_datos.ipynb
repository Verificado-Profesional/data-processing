# **1. Preparación de dataframes**

**Objetivos de la sección:**

*   Setear nombres de columnas

*   Setear las clasificaciones en *True* o *Fake* según corresponda

*   Eliminar columnas sobrantes







```python
db1 = pd.read_csv('/content/drive/MyDrive/Trabajo Profesional/Bases de datos/1-DB1-VF.csv', sep=';', names=['Category', 'Headline', 'Text', 'Date'], header=0)
db1['Category'] = db1['Category'].replace({1: 'True', 0: 'Fake'})
db1
```

```python
db2 = pd.read_excel('/content/drive/MyDrive/Trabajo Profesional/Bases de datos/2-DB2-VF.xlsx')
db2 = db2.drop(columns=['Id'])
db2
```

```python
db3 = pd.read_csv('/content/drive/MyDrive/Trabajo Profesional/Bases de datos/3-DB3-VF.csv', header=0, names=['Category', 'Text'])
db3['Category'] = db3['Category'].replace({True: 'True', False: 'Fake'})
db3
```

```python
db4 = pd.read_csv('/content/drive/MyDrive/Trabajo Profesional/Bases de datos/4-DB4-VF.csv', header=0, names=['Text', 'Category'])
db4['Category'] = db4['Category'].replace({'fake': 'Fake', 'real': 'True'})
db4
```

```python
db5 = pd.read_csv('/content/drive/MyDrive/Trabajo Profesional/Bases de datos/5-DB5-VF.csv', header=0, names=['Headline', 'Text', 'Category'])
db5['Category'] = db5['Category'].replace({1: 'True', 0: 'Fake'})
db5
```

# **2. Preparación dataframe final**

**Objetivos de la sección:**

*   Unificar dataframes en uno

*   Eliminar columnas sobrantes

*   Datos faltantes como NaN

```python
DB = db1.append(db2.append(db3.append(db4.append(db5)))).reset_index()
DB = DB.drop(columns=['index'])
DB
```

# **3. Análisis de datos**

```python
import matplotlib.pyplot as plt
```

**Cantidad de nulos por columna:**

```python
DB.isnull().sum()
```

**Cantidad de noticias Verdaderas y Falsas:**

```python
counter_true = DB[DB['Category'] == 'True']['Category'].count()
counter_fake = DB[DB['Category'] == 'Fake']['Category'].count()
counter = [counter_true, counter_fake]

categories = ['True', 'Fake']
```

```python
# Crea un gráfico de barras
plt.bar(categories, counter)

# Agrega un título al gráfico
plt.title("Cantidad de noticias verdaderas y falsas")

# Agrega etiquetas a los ejes
plt.xlabel("Categorías")
plt.ylabel("Cantidad de noticias")

# Muestra el gráfico
plt.show()
```

**Cantidad de noticias por tópico**

```python
count_per_topic = DB.groupby(DB['Topic']).size().reset_index(name="Count")
count_per_topic
```

```python
plt.pie(count_per_topic["Count"], labels=count_per_topic["Topic"], autopct="%1.1f%%")

# Agrega un título al gráfico
plt.title("Cantidad de noticias por tópico")

# Muestra el gráfico
plt.show()
```

**Cantidad de noticias por fuente**

```python
count_per_source = DB.groupby(DB['Source']).size().reset_index(name="Count")
count_per_source
```

```python
plt.figure(figsize=(12, 30))  # Ajusta el tamaño según sea necesario
DB.groupby('Source').size().sort_values().plot(kind='barh')
plt.title('Cantidad de Noticias por Fuente')
plt.xlabel('Número de Noticias')
plt.ylabel('Fuente')
plt.show()
```

**Cantidad de noticias y su categoria por año**

```python
DB['Date'] = pd.to_datetime(DB['Date'])

# Obtener la fecha más antigua y la fecha más reciente
fecha_mas_antigua = DB['Date'].min()
fecha_mas_reciente = DB['Date'].max()

[fecha_mas_antigua, fecha_mas_reciente]
```

```python
DB['Date'] = pd.to_datetime(DB['Date'])

# Obtener el año de cada noticia
DB['Year'] = DB['Date'].dt.year

# Graficar la cantidad de noticias por año y clasificación
plt.figure(figsize=(12, 6))
DB.groupby(['Year', 'Category']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Cantidad de Noticias por Año y Clasificación')
plt.xlabel('Año de Publicación')
plt.ylabel('Número de Noticias')
plt.legend(title='Clasificación')
plt.show()
```
